```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import plotly.express as px
```

```python
olympic_medals = pd.read_csv('/content/drive/MyDrive/DATA/olympic_medals.csv')
```

```python
olympic_medals = olympic_medals[olympic_medals['Year'] >= 2000]
```

```python
olympic_medals['Medal'] = olympic_medals['Medal'].replace({'G':'Gold', 'S':'Silver', 'B':'Brown'})
```

```python
fig = px.parallel_categories(olympic_medals,
                             dimensions=['Gender', 'Medal'],
                             color='Medal',
                             color_continuous_scale=px.colors.sequential.Greys)
```

```python
fig.show()
```
